```python
#Importando Bibliotecas
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
#Importando as tábuas com os qx
qx_morta = pd.read_csv(str(Path(os.getcwd())) + '\\Tábuas\\tb_mortalidade.csv', encoding='UTF-8', sep=';')
qx_morta_inv = pd.read_csv(str(Path(os.getcwd())) + '\\Tábuas\\tb_mortalidade_inv.csv', encoding='UTF-8', sep=';')
qx_entrada_inv = pd.read_csv(str(Path(os.getcwd())) + '\\Tábuas\\tb_entrada_inv.csv', encoding='UTF-8', sep=';')
qx_morbidez = pd.read_csv(str(Path(os.getcwd())) + '\\Tábuas\\tb_morbidez.csv', encoding='UTF-8', sep=';')
```

```python
#Calculando os px
px_morta = (1 - qx_morta)
px_morta_inv = (1 - qx_morta_inv)
px_entrada_inv = (1 - qx_entrada_inv)
px_morbidez = (1 - qx_morbidez)
```

```python
#Calculando os dx e lx
dx_morta = pd.DataFrame(columns=qx_morta.columns)
dx_morta_inv = pd.DataFrame(columns=qx_morta_inv.columns)
dx_entrada_inv = pd.DataFrame(columns=qx_entrada_inv.columns)
dx_morbidez = pd.DataFrame(columns=qx_morbidez.columns)

lx_morta = pd.DataFrame(columns=qx_morta.columns)
lx_morta_inv = pd.DataFrame(columns=qx_morta_inv.columns)
lx_entrada_inv = pd.DataFrame(columns=qx_entrada_inv.columns)
lx_morbidez = pd.DataFrame(columns=qx_morbidez.columns)

for c in range(127):
    if c >= 1:
        lx_morta.loc[c] = lx_morta.loc[c-1] - dx_morta.loc[c-1]
        lx_morta_inv.loc[c] = lx_morta_inv.loc[c-1] - dx_morta_inv.loc[c-1]
        lx_entrada_inv.loc[c] = lx_entrada_inv.loc[c-1] - dx_entrada_inv.loc[c-1]
        lx_morbidez.loc[c] = lx_morbidez.loc[c-1] - dx_morbidez.loc[c-1]

        dx_morta.loc[c] = lx_morta.loc[c] * qx_morta.loc[c]
        dx_morta_inv.loc[c] = lx_morta_inv.loc[c] * qx_morta_inv.loc[c]
        dx_entrada_inv.loc[c] = lx_entrada_inv.loc[c] * qx_entrada_inv.loc[c]
        dx_morbidez.loc[c] = lx_morbidez.loc[c] * qx_morbidez.loc[c]

    else:
        lx_morta.loc[c] = 100000
        lx_morta_inv.loc[c] = 100000
        lx_entrada_inv.loc[c] = 100000
        lx_morbidez.loc[c] = 100000

        dx_morta.loc[c] = lx_morta.loc[c] * qx_morta.loc[c]
        dx_morta_inv.loc[c] = lx_morta_inv.loc[c] * qx_morta_inv.loc[c]
        dx_entrada_inv.loc[c] = lx_entrada_inv.loc[c] * qx_entrada_inv.loc[c]
        dx_morbidez.loc[c] = lx_morbidez.loc[c] * qx_morbidez.loc[c]
```
